3 - Crie uma nova coluna chamada ‘’tipo_faixa_etaria”, onde deverá conter os seguintes valores:

 - Crianças: 3 a 12 anos

 - Adolescentes: 13 a 19 anos

 - Jovens adultos: 20 a 39 anos 

 - Adultos de meia-idade: 40 a 59 anos 

 - Idosos: 60 anos ou mais

```python
def categorizar_faixa_etaria(idade):
    if 3 <= idade <= 12:
        return 'Crianças: 3 a 12 anos'
    elif 13 <= idade <= 19:
        return 'Adolescentes: 13 a 19 anos'
    elif 20 <= idade <= 39:
        return 'Jovens adultos: 20 a 39 anos'
    elif 40 <= idade <= 59:
        return 'Adultos de meia-idade: 40 a 59 anos'
    elif idade >= 60:
        return 'Idosos: 60 anos ou mais'
    else:
        return 'Outra faixa etária'

# Aplicar a função à coluna 'idade' e criar a nova coluna 'tipo_faixa_etaria'
df['tipo_faixa_etaria'] = df['idade'].apply(categorizar_faixa_etaria)
df.head(5)
```

Crie uma nova coluna chamada ‘’regiao”, onde deverá conter valores:

 - Metropolitana: Recife, Olinda, Jaboatão dos Guararapes, Paulista, Camaragibe, São Lourenço da Mata, Abreu e Lima, Igarassu, Cabo de Santo Agostinho

 - Interior: Cidades que não estão inclusas em Metropolitana


```python
regiao_metropolitana =  ['RECIFE', 'OLINDA', 'JABOATÃO DOS GUARARAPES', 'PAULISTA', 
                        'CAMARAGIBE', 'SÃO LOURENÇO DA MATA', 'ABREU E LIMA', 
                        'IGARASSU', 'CABO DE SANTO AGOSTINHO']

# Função para categorizar a região
def categorizar_regiao(municipio):
    if municipio  in  regiao_metropolitana:
        return 'Metropolitana'
    else:
        return 'Interior'

# Aplicar a função à coluna 'municipio' e criar a nova coluna 'regiao'
df['regiao'] = df['municipio'].apply(categorizar_regiao)

# Visualizar o DataFrame com a nova coluna
df.head(5)
```

Conte quantas pessoas receberam a vacina para cada cada  ’regiao” e cada ‘’tipo_faixa_etaria”

```python
contagem_por_regiao_faixa_etaria = df.groupby(['regiao', 'tipo_faixa_etaria']).size().reset_index(name='quantidade')

print(contagem_por_regiao_faixa_etaria)
```

Determine a idade média das pessoas vacinadas para  cada  ’regiao” e cada ‘’tipo_faixa_etaria”

```python
media_de_idade_pessoas_vacinadas = df.groupby(['regiao', 'tipo_faixa_etaria'])['idade'].mean().reset_index()

print(media_de_idade_pessoas_vacinadas)
```

Identifique e exiba o tipo de vacina mais administrado para cada  ’regiao” e cada ‘’tipo_faixa_etaria”

```python
media_vacina_usada = df.groupby(['regiao', 'tipo_faixa_etaria'])['vacina_fabricante'].agg(lambda x: x.mode()[0]).reset_index()

print(media_vacina_usada)
```
